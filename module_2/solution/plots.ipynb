# Подсчет статистики

- количество публикаций об упоминании университета
- количество публикующих контент пользователей
- количество лайков/просмотров/комментариев/репостов
- график количество публикаций в день за собираемый период

```python
import plotly.express as px
import pandas as pd

msu_df = pd.read_csv('../data/msu_september.csv')
spbgu_df = pd.read_csv('../data/spbgu_september.csv')

df = pd.concat([spbgu_df, msu_df])
```

```python
agg_df = df.groupby(['topic', 'date']).agg({ 'from_id': 'nunique',
                                             'id': 'nunique',
                                             'views': 'sum',
                                             'likes': 'sum',
                                             'comments': 'sum',
                                             'reposts': 'sum'}).reset_index()
agg_df = agg_df[agg_df['date'] < '2024-09-09']
agg_df
```

### Статистические показатели за весь период

```python
agg_df.groupby('topic').aggregate({'id': 'sum',
                                   'from_id': 'sum',
                                   'views':'sum',
                                   'likes': 'sum',
                                   'comments': 'sum',
                                   'reposts': 'sum'}).reset_index()
```

```python
metrics = {'views': 'просмотров', 'likes': 'лайков', 'comments': 'комментариев',
           'reposts': 'репостов', 'from_id': 'публикующих пользователей', 'id': 'публикаций'}
custom_colors = ['#023047', '#ffb703']

for metric in metrics.keys():
    metric_df = agg_df.groupby('topic')[metric].sum().reset_index()
    
    # Bar Plot
    fig_bar = px.bar(metric_df, x='topic', y=metric, title=f'Количество {metrics[metric]}', 
                     labels={'topic': 'Tag', metric: f'Количество {metrics[metric]}'},
                     color='topic', color_discrete_sequence=custom_colors)
    fig_bar.show()
    
    # Pie Plot
    fig_pie = px.pie(metric_df, names='topic', 
                     values=metric, title=f'Количество {metrics[metric]}',
                     color='topic', color_discrete_sequence=custom_colors)
    fig_pie.show()
```

```python
fig_line = px.line(agg_df, x='date', y='id', color='topic', 
                   title='Количество публикаций по дням',
                   labels={'date': 'Дата', 'id': 'Число публикаций'},
                   color_discrete_sequence=custom_colors)
fig_line.show()
```
